```python
# Matrix Multiplication
a1 = np.array([[1, 3],
               [0, 1],
               [6, 2],
               [9, 7]])
 
b1 = np.array([[4, 1, 3],
               [5, 8, 5]])
```

```python
# Manipulating images
from PIL import Image
import numpy as np

image = Image.open('yummy_macarons.jpg')
img_array = np.array(image)
plt.imshow(img_array)
```

```python
type(img_array)
```

```python
img_array.shape
```

```python
img_array.ndim
```

```python
sRGB_array = img_array / 255
```

```python
grey_vals = np.array([0.2126, 0.7152, 0.0722])
```

```python
img_gray = sRGB_array @ grey_vals
```

```python
img_gray = sRGB_array @ grey_vals
img_gray = np.matmul(sRGB_array, grey_vals)
```

```python
plt.imshow(img_gray, cmap='gray')
```

```python
plt.imshow(img_gray)
```

```python
# manipulating the nd-array to change the image
plt.imshow(np.flip(img_gray), cmap='gray')
```

```python
plt.imshow(np.rot90(img_array))
```

```python
solar_img = 255 - img_array
plt.imshow(solar_img)
```
